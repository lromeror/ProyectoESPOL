```python
#Funcion Replace Media
#bore
#house 
def replace_Media(nameData,feature):
  mean_value=nameData[feature].astype('float').mean()
  nameData[feature].replace(np.nan,mean_value,inplace=True)
  
```

```python
list_feature=['bore','stroke','horsepower','peak-rpm']
for fea in list_feature:
  replace_Media(df_automobile2,fea)

df_automobile2.isnull().sum()
```
