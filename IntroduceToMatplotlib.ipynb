```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Figure and Axes Objects

```python
#The Figure object in Matplotlib represents the entire figure or window where the plot is drawn
#Create a figure object
fig=plt.figure()
```

```python
#Create an Axes object
ax=fig.add_subplot(111)#Recibe 3 parametros
#111 means one row, one column, and plot number 1
#three arguments: the number of rows, number of columns, and the plot number sin comas
fig
```

```python
#Generate a some data
x = np.linspace(0,12.57,200)


y = np.sin(x)#Esta es la funcion seno 

#Plot data 
#es como que anadimos la info de plot
ax.plot(x,y)

fig
```

```python
# Add axis labels and a title
ax.set_xlabel('x')#Para anadir etiquetas
ax.set_ylabel('sin(x)')
ax.set_title('A simple Plot of SIN')

fig
```

```python
ar2=np.linspace(0,10,100)
#linspace solo necesita especificar la cantidad de puntos en el intervalo, sin preocuparse por el tamaño del paso
#porner numero de incio, fin y cuantos numeros partidos de forma igual
print(ar2)
#np.linspace(start, stop, num, endpoint, retstep, dtype, axis)

#num es el número de puntos espaciados uniformemente. Y es un parámetro opcional con un valor predeterminado de 50.
#Endpoint también es un parámetro opcional que puede ser Verdadero o Falso.

#El valor predeterminado es Verdadero, lo que significa que el punto final se incluirá en el intervalo de forma predeterminada. 
#Sin embargo, puede establecerlo en False para excluir el punto final.
#retstep es otro parámetro opcional que toma los valores booleanos True o False.
#Cuando se establece en True, se devuelve el valor del paso. como ha venido saltando para estar iguales
```
