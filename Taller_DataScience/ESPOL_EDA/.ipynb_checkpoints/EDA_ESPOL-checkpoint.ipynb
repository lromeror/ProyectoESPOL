```python
li = []
for x in list(df.HORAINICIO):
    li_ho = x.split(' ')
    if 'a' in li_ho[1]:
        li_t = li_ho[0].split(':')
        if len(li_t[0]) == 1:
            n = '0'+li_ho[0] + ' AM'
            li.append(n)
        elif len(li_t[0]) > 1:
            n = li_ho[0] + ' AM'
            li.append(n)

    elif 'p' in li_ho[1]:
        li_t = li_ho[0].split(':')
        if 1 == len(li_t[0]):
            n = '0' + li_ho[0] + ' PM'
            li.append(n)
        elif len(li_t[0]) > 1:
            n = li_ho[0] + ' PM'
            li.append(n)


df['HORAINICIO'] = li
```

```python
df
```

```python
from datetime import datetime
FinHo = []
for x in df['HORAINICIO']:
    x = str(x)
    forma = '%I:%M:%S %p'
    n = datetime.strptime(x, forma).time()
    FinHo.append(n)

df['HORAINICIO'] = FinHo
```

```python
#Pasar a int las columnas
def cambiarPunt(df,fea):
    w=0
    for i in df[fea]:
        rem = int(i.split(',')[0])
        df.loc[w,fea] = rem#Es para indexar por su ID, en este caso el ID es numeros
        w+=1
    return df[fea]

for name_col in ['PARALELO','CAPACIDAD','CUPOPLANIFICADO']:
  df[name_col] = cambiarPunt(df,name_col)
```
