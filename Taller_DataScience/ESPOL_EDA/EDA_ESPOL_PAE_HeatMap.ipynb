```python
def convert_time(df,feature):
    li = []
    for x in list(df[feature]):
        li_ho = x.split(' ')
        if 'a' in li_ho[1]:
            li_t = li_ho[0].split(':')
            if len(li_t[0]) == 1:
                n = '0'+li_ho[0] + 'AM'
                li.append(n)
            elif len(li_t[0]) > 1:
                n = li_ho[0] + 'AM'
                li.append(n)

        elif 'p' in li_ho[1]:
            li_t = li_ho[0].split(':')
            if 1 == len(li_t[0]):
                n = '0' + li_ho[0] + 'PM'
                li.append(n)
            elif len(li_t[0]) > 1:
                n = li_ho[0] + 'PM'
                li.append(n)

    df[feature] = li
```

```python
def normalizar_lon(df,bloque,lon):
    modiCor=[]
    compa=[]
    cont=0
    for bloque,lo in zip (df[bloque],df[lon]):
        if bloque not in compa:
            compa.append(bloque)
            modiCor.append(lo)
        else:
            cont=cont+0.000009
            
            modiCor.append(lo-cont)
    
    df[lon+'2']=modiCor
```

```python
def normalizar_lat(df,bloque,lat):
    modiCor=[]
    compa=[]
    cont=0
    for bloque,lo in zip (df[bloque],df[lat]):
        if bloque not in compa:
            compa.append(bloque)
            modiCor.append(lo)
        else:
            cont=cont+0.0000009
            
            modiCor.append(lo-cont)
    
    df[lat+'2']=modiCor
```

```python
def dia_ord(df):
    # Asigna un valor asociado de 0, 1, 2 o 3 a los d√≠as lunes, martes, miercoles y jueves respectivamente. 
    # Esto se recoge en un nueva columna para el dataframe
    L = []
    for dia in df.DIA:
        if dia == 'lunes':
            L.append(0)
        elif dia == 'martes':
            L.append(1)
        elif dia == 'mi√©rcoles':
            L.append(2)
        else:
            L.append(3)
    df['AUX_DIA'] = L
```

```python
def create_list(value):
    #Crea una lista de listas, la cantidad de listas que esten dentro de la lista depende del valor value
    L = []
    for i in range(value):
        L.append([0, 0, 0, 0])
    return L
```

```python
def crear_lista_heatmap(df_blo):
    # Toma un dataframe que ha sido segreadado por el nombre de bloques de los edificios y devuelve una lista de listas
    # donde cada lista representa a las filas del heatmap, y los elementos dentro de las listas a la posici√≥n de las columnas y
    # al valor de cada celda. 
    df_bloque = df_blo.copy()
    dia_ord(df_bloque)
    df_bloque = df_bloque.sort_values('HORAINICIO')
    l_hora = list(df_bloque.HORAINICIO.unique())
    total_of_lists = len(l_hora)
    L = create_list(total_of_lists)
    for i in range(total_of_lists):
        value = l_hora[i]
        df_blo_h = df_bloque[df_bloque.HORAINICIO == value].sort_values('AUX_DIA')
        if (len(list(df_blo_h.AUX_DIA))) != 4:
            for aux, cupo in zip(list(df_blo_h.AUX_DIA), list(df_blo_h.CUPOPLANIFICADO)):
                if aux == 0:
                    L[i][aux] = cupo
                elif aux == 1:
                    L[i][aux] = cupo
                elif aux == 2:
                    L[i][aux] = cupo
                elif aux == 3:
                    L[i][aux] = cupo
                else:
                    L[i][aux] = 0
        else:
            L[i] = list(df_blo_h.CUPOPLANIFICADO)       
    return L
```

```python
def crear_etiquetas_heatmap(df_blo):
    # Crea las etiquetas para los valores asociados a la lista creada con la funcion anterior.
    lista=[]
    df_blo2 = df_blo.sort_values('HORAINICIO')
    for i in df_blo2['HORAINICIO']:
        i=str(i)
        li=i.strip().split(':')
        lista.append(str(li[0])+':'+str(li[1]))
    df_blo2['ETIQUETAHORA'] = lista
    etiq = list(df_blo2.ETIQUETAHORA.unique())
    return  etiq
```

```python
def crear_lista_heatmap_fac(df_blo):
    df_bloque = df_blo.copy()
    dia_ord(df_bloque)
    df_bloque = df_bloque.sort_values('HORAINICIO')
    l_hora = list(df_bloque.HORAINICIO.unique())
    total_of_lists = len(l_hora)
    L = create_list(total_of_lists)
    L_title = []
    for i in range(total_of_lists):
        value = l_hora[i]
        df_blo_h = df_bloque[df_bloque.HORAINICIO == value].sort_values('AUX_DIA')
        if (len(list(df_blo_h.AUX_DIA))) != 4:
            for aux, cupo in zip(list(df_blo_h.AUX_DIA), list(df_blo_h.CUPOPLANIFICADO)):
                if aux == 0:
                    L[i][aux] = cupo
                elif aux == 1:
                    L[i][aux] = cupo
                elif aux == 2:
                    L[i][aux] = cupo
                elif aux == 3:
                    L[i][aux] = cupo
                else:
                    L[i][aux] = 0
        else:
            L[i] = list(df_blo_h.CUPOPLANIFICADO)       
    return L
```

```python
def normalizarDuracion(df,duracion):
    lista=[]
    for i in df[duracion]:
        i=str(i)
        li=i.strip().split(':')
        li=list(np.array(li,dtype='int'))
        if li[1]==0:
            lista.append(li[0])
        elif li[1]!=0:
            lista.append(li[0]+0.5)
    
    df[duracion+' 2']=lista
```
