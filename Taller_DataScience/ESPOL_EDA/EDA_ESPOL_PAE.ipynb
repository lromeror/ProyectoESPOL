```python
def convert_time(df,feature):
    li = []
    for x in list(df[feature]):
        li_ho = x.split(' ')
        if 'a' in li_ho[1]:
            li_t = li_ho[0].split(':')
            if len(li_t[0]) == 1:
                n = '0'+li_ho[0] + 'AM'
                li.append(n)
            elif len(li_t[0]) > 1:
                n = li_ho[0] + 'AM'
                li.append(n)

        elif 'p' in li_ho[1]:
            li_t = li_ho[0].split(':')
            if 1 == len(li_t[0]):
                n = '0' + li_ho[0] + 'PM'
                li.append(n)
            elif len(li_t[0]) > 1:
                n = li_ho[0] + 'PM'
                li.append(n)

    df[feature] = li
```
