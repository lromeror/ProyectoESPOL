```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random as rd
```

```python
df=pd.read_csv('country_profile_variables_taller1.csv')
df
```

```python
list_id=list(range(df.shape[0]+1))
id_ale=rd.sample(list_id,20)

df_new=df.iloc[id_ale]
df_new=df_new[['Energy production, primary (Petajoules)','CO2 emission estimates (million tons/tons per capita)','Individuals using the Internet (per 100 inhabitants)','Seats held by women in national parliaments %','Urban population (% of total population)','Agricultural production index (2004-2006=100)','Economy: Services and other activity (% of GVA)','Economy: Industry (% of GVA)','GDP per capita (current US$)','GDP: Gross domestic product (million current US$)','Sex ratio (m per 100 f, 2017)','Population density (per km2, 2017)','Population in thousands (2017)']]

df_new
```

```python
def rangoAmpl(Feature,CanIntervalos,name):
  max=Feature.max()
  min=Feature.min()
  rg=max-min
  ampl=rg/CanIntervalos
  intervals = pd.interval_range(
    start=min,
    end=max+ampl,
    freq=ampl,
    name=name,
    closed="left"
    )
  
  dataFre=pd.DataFrame(index=intervals)
  dataFre['Marca']=dataFre.index.mid
  dataFre['Fre_Abs']=pd.cut(Feature, bins=dataFre.index).value_counts()

  li_FeAcumu=[]
  acumu=0
  for x in dataFre['Fre_Abs']:
    acumu=acumu+x
    li_FeAcumu.append(acumu)

  dataFre['Fre_Acumu']=li_FeAcumu

  CantidadDatos=dataFre['Fre_Abs'].sum()

  dataFre['Fre_Re']=dataFre['Fre_Abs']/CantidadDatos

  li_Fe_reAcumu=[]

  acumu2=0
  for i in dataFre['Fre_Re']:
    acumu2=acumu2+i
    li_Fe_reAcumu.append(acumu2)

  dataFre['Fre_Rela_Acumu']=li_Fe_reAcumu
  
  return dataFre
```

```python
data1=rangoAmpl(df_new['Population in thousands (2017)'],20,'Population in thousands (2017)')
data1
```

```python
data2=rangoAmpl(df_new['Population density (per km2, 2017)'],20,'Population density (per km2, 2017)')
data2
```

```python
col=list(df_new.columns.values)
for k in col:
  data=rangoAmpl(df_new[k],20,k)
  print(data.to_markdown())
```
